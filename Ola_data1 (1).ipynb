```python
import pandas as pd
```

```python
df = pd.read_csv("Bookings-70000-Rows.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
if df['V_TAT'].isnull().any() :
  df['V_TAT'] = df['V_TAT'].fillna(0);
```

```python
if df['C_TAT'].isnull().any() :
  df['C_TAT'] = df['C_TAT'].fillna(0);
```

```python
if df['Incomplete_Rides'].isnull().any():
  df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna('Yes')
```

```python
df['Final_Payment'] = df['Booking_Value']
```

```python
df.loc[df['Incomplete_Rides'] == 'Yes', 'Final_Payment'] = 0
```

```python
if df['Payment_Method'].isnull().any():
  df['Payment_Method'] = df['Payment_Method'].fillna('Not applicable')
```

```python
if df['Incomplete_Rides_Reason'].isnull().any():
  df['Incomplete_Rides_Reason'] = df['Incomplete_Rides_Reason'].fillna('No')
```

```python
df.isnull().sum()
```

```python
if df['Canceled_Rides_by_Customer'].isnull().any():
  df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna('No')
```

```python
if df['Canceled_Rides_by_Driver'].isnull().any():
  df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna('No')
```

```python
if df['Driver_Ratings'].isnull().any():
  df['Driver_Ratings'] = df['Driver_Ratings'].fillna('NA')
```

```python
if df['Customer_Rating'].isnull().any():
  df['Customer_Rating'] = df['Customer_Rating'].fillna('NA')
```

```python
df.isnull().sum()
```
